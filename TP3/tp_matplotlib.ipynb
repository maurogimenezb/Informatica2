```python
import numpy as np
import matplotlib.pyplot as plt
```

# 1. Graficar una señal sinusoidal en función del tiempo:

```python
# Generar dos arrays de numpy con 100 puntos cada uno

t = np.linspace(0, 10, 100)  # Array de tiempo
a = np.sin(t)                # Array de amplitud
```

```python
# Crear la gráfica

plt.figure(figsize=(10, 6))
plt.plot(t, a)
plt.xlabel('Tiempo')
plt.ylabel('Amplitud')
plt.title('Señal sinusoidal en función del tiempo')
plt.grid(True)
plt.show()
```

# 2. Crear un histograma de datos aleatorios:

```python
# Generar un array de numpy con 1000 números aleatorios siguiendo una distribución normal

data = np.random.normal(size=1000)
```

```python
# Crear el histograma

plt.figure(figsize=(10, 6))
plt.hist(data, bins=30, edgecolor='black')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.title('Histograma de datos aleatorios')
plt.grid(True)
plt.show()
```

# 3. Crear un gráfico de dispersión de puntos:


```python
# Generar dos arrays de numpy con 100 puntos cada uno que representen las coordenadas x e y

x = np.random.rand(100)
y = np.random.rand(100)
```

```python
# Crear el gráfico de dispersión

plt.figure(figsize=(10, 6))
plt.scatter(x, y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Gráfico de dispersión de puntos')
plt.grid(True)
plt.show()
```

# 4. Crear un gráfico de superficie 3D de una función z = f(x, y):

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Generar arrays de coordenadas x e y

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
```

```python
# Definir la función z = f(x, y)

z = np.sin(np.sqrt(x**2 + y**2))
```

```python
# Crear el gráfico de superficie 3D

fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='viridis')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Gráfico de superficie 3D')
plt.show()
```
