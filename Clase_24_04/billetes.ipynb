## Modelo de Clasifiacion

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn import svm
from sklearn.linear_model import Perceptron
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
```

```python
# Lectura de datos

df = pd.read_csv('banknotes.csv')
df.head()
```

```python
df.describe()
```

```python
plt.scatter (df.iloc [:,0], df.iloc[:,1])
plt.xlabel (df.columns[0])
plt.ylabel (df.columns[1])
```

```python
legales = df.loc [df['class']==1]
ilegales =df.loc [df['class']==0]
```

```python
fig, axs = plt.subplots (nrows=2, ncols=3, figsize=(10,10))

for i, ax in enumerate(axs):
    ax.scatter (legales.iloc [:,0], legales.iloc [:,i+1], label ='legal' )
    ax.scatter (legales.iloc [:,0], ilegales.iloc [:,i+1], label ='ilegal' )
    ax.set_ylabel (df.columns[0])
    ax.set_xlabel (df.columns[i+1])
    ax.legend()
```

# Modelado 

```python
X = df.iloc [:,:-1]
y = df.iloc[:,-1]

y.head()
```

```python
#Separamos los datos de entrenamiento de prediccion o verificacion

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.26, random_state=50)
```

```python
#Adoptamos un modelo
model = svm.SVC()
```

```python
#Entrenamiento

model.fit(X_train, y_train)
```

```python
# Predicciones 
preds = model.predict  (X_test)
score = model.score (X_test, preds)
score
```
