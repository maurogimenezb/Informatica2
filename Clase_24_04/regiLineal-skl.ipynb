# Modelo de regresion lineal

```python
#Entradas
X_total = df ['Total'].to_numpy().reshape(-1,1)

#Salida
y_total = df ['Sales']
```

```python
Xtotal_train, Xtotal_test, ytotal_train, ytotal_test = train_test_split (X_total, y_total, test_size=0.25, random_state=100)
```

```python
model = LinearRegression()
model.fit(Xtotal_train, ytotal_train)
w_total = model.coef_ #Pendiente de la recta
b_total = model.intercept_ #Interseccion de la recta con el eje y
## y = mx + b



```

```python
#Predicciones

total_pred = np.dot(Xtotal_test, w_total) + b_total #Ecuacion de la recta
s_total = model.score(Xtotal_test, ytotal_test) #Bondad de ajuste R cuadrado 2

s_total

#Prediccion para un valor particular
pred_particular = np.dot (100, w_total) + b_total
pred_particular
```

```python
plt.scatter (Xtotal_test, ytotal_test)
plt.plot(Xtotal_test, total_pred, 'red')

#FALTA CODIGO LE SAQUE FOTO



plt.legend ()

plt.title ('')
plt.xlabel ('')
plt.ylabel ('')
plt.grid ()
```
